```python
#importar as bibliotecas
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#definir as entradas para um problema que temos um grafico tamanho em função do preço
x_tamanho = np.array([1.0, 2.0])
y_preco = np.array([300.0, 500.0])
print(f"x = {x_tamanho}")
print(f"y = {y_preco}")
```

```python
#aqui temos que definir m como sendo o numero total de treinamentos
m = len(x_tamanho)
print(f"numero de treinamentos eh: {m}")
```

```python
#para definir o enésino(th) treinamento de exemplo:
i = 0
x_i = x_tamanho[i]
y_i = y_preco[i]
print(f"(xi,yi) = ({x_i}, {y_i})")
```

```python
#plotando os dados:
#marca os pontos com x e define a cor como vermelha 
plt.scatter(x_tamanho, y_preco, marker = 'o', c = 'r')
plt.title("Preço das casas")
plt.ylabel("Preço (em mil)")
plt.xlabel("Tamanho em m2")
plt.show
```

```python
#definimos um w e um b aleatorios
w = 200
b = 100
```

```python
#sabemos que a função de regressão linear eh f(w,b) = wx + b, definindo uma função para ele
def modelo_output(x,y,b):
    m = len(x_tamanho)
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i] = w * x[i] + b
    return f_wb
      
```

```python
#chamando a funçaão e plotando a saída:
call_f_wb = modelo_output(x_tamanho, w, b)
plt.plot(x_tamanho, call_f_wb, c = 'b', label = 'previsao')
plt.scatter(x_tamanho, y_preco, marker = 'o', c = 'r', label = 'dados')
plt.title("Preço das casas")
plt.ylabel("Preço (em mil)")
plt.xlabel("Tamanho em m2")
plt.legend()
plt.show
```
